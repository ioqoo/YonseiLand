```python
from get_mileage_result import *
from analysis import *
import matplotlib
import matplotlib.pyplot as plt
```

```python
file_list = get_file_list()
scatter_nonmajor = []
scatter_major = []
```

```python
for file in file_list:
    Class_Info, Mileage_Result = load_data_pd(default_path_dir + file)
    try:
        min_mileage_nonmajor, min_mileage_major = train(Class_Info, Mileage_Result)
        quota = Class_Info[''][0]
        participants = Class_Info[''][0]
        if (quota == 0): continue
        if quota < participants:
            scatter_nonmajor.append([participants/quota, min_mileage_nonmajor])
            scatter_major.append([participants/quota, min_mileage_major])
    except:
        continue
```
